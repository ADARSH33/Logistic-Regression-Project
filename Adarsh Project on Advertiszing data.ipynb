```python
#now we need to perform the logistic regression for predicting the CLICKED ON AD
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = a_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = a_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=40)
```

```python
#training and fitting the logistic model on the training set
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticmodel=LogisticRegression()
```

```python
logisticmodel.fit(X_train,y_train)
```

```python
#now we need to predict and evaluate
```

```python
predicting=logisticmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predicting))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predicting))
```
